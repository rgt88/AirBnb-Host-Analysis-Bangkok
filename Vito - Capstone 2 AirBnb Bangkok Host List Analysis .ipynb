2. ***Data Understanding***<br>

This dataset have 16 columns that give an information about traffic, employee / hoster, and area Airbnb in Bangkok Thailand.
There are the columns and the meanings of each columns: <br>
1. id =  Airbnb's unique identifier for the listing. <br>
2. name = Name of the listing.<br>
3. host_id = Airbnb's unique identifier for the host/user. <br>
4. host_name = Name of the host. Usually, just the first name(s).<br>
5. neighborhood = The neighborhood is geocoded using the latitude and
longitude against neighborhoods as defined by open or public
digital shapefiles.<br>
6. latitude = Uses the World Geodetic System (WGS84) projection for
latitude and longitude.<br>
7. Longitude = Uses the World Geodetic System (WGS84) projection for
latitude and longitude.<br>
8. room_type = [Entire home/apt |Private room| Shared room| Hotel] <br>

Entire Room : Entire places are best if you're seeking a home away from
home. With an entire place, you'll have the whole space to
yourself. This usually includes a bedroom, a bathroom, a
kitchen, and a separate, dedicated entrance. Hosts should
note in the description if they'll be on the property or not (ex:
"Host occupies the first floor of the home") and provide further
details on the listing.<br>

Private Room: Private rooms are great for when you prefer a little privacy,
and still value a local connection. When you book a private
room, you'll have your private room for sleeping and may
share some spaces with others. You might need to walk
through indoor spaces that another host or guest may occupy
to get to your room.<br>

Shared Room: Shared rooms are for when you don't mind sharing a space
with others. When you book a shared room, you'll be sleeping
in a space that is shared with others and share the entire
space with other people. Shared rooms are popular among
flexible travelers looking for new friends and budget-friendly stays<br>

9. price = Daily price in local currency. Note, the $ sign may be used
despite the locale.<br>
10. minimum_nights = The minimum number of night stays for the listing (calendar
rules may differ).<br>
11. number_of_reviews = The number of reviews the listing has.<br>
12. last_review = The date of the last/newest review.<br>
13. review_per_month = The month of the last review on month <br>
14. calculated_host_listings_count = The number of listings the host has in the current scrape in
the city/region geography.<br>
15. availability_365 = avaliability_x. The calendar determines the availability of the
listing x days in the future. Note a listing may be available
because it has been booked by a guest or blocked by the
host.<br>
16. number_of_reviews_itm = The number of reviews the listing has (in the last 12 months).<br>

Statistic Inferensia

```python
from scipy.stats import kruskal
```

H0: There is no significant difference between price and room type on Airbnb.<br>
H1: There is a significant difference between price and room type on Airbnb.<br>

```python
# separate data into independent variable (harga) and dependent variable (tipe kamar)
tipe_kamar = df1["room_type"]
harga = df1["price"]

# perform Kruskal-Wallis test
result = kruskal(*[harga[tipe_kamar == tipe] for tipe in tipe_kamar.unique()])

# print results
print("Kruskal-Wallis H-test")
print("H-statistic:", result.statistic)
print("p-value:", result.pvalue)
```

Interpretasi: We get p-value < 0.05, the meaning is there is a significant difference between price and room type on Airbnb. So more expensive price will get higher classification of room type

Room Type Hotel Room is the most expensive type room based on median

4. What is the most name listing in Airbnb bangkok? <br>

```python
# Analyzing the words used by hosts when naming their listings
# getting list of all names
names = df1['name'].tolist()
name_words = []
for name in names:
    name = str(name).split()
    for word in name:
        name_words.append(word.lower())
```

```python
import nltk # natural language toolkit
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

filtered_words = [word for word in name_words if word not in stopwords.words('english')]
print(filtered_words[:25])
```

```python
from collections import Counter # to count words in our list

# plotting top 25 words used by the host in naming their home
words_count = Counter(filtered_words).most_common()
words_count = words_count[:25]
```

```python
# converting the data into a dataframe so we can plot it using Seaborn
words_df = pd.DataFrame(words_count)
words_df.head(10)
```

```python
words_df.rename(columns={0:"Words", 1:"Count"}, inplace=True)
```

```python
plt.figure(figsize=(10,5))
fig, ax = plt.subplots()
fig.suptitle('Top 25 Words Used in Bangkok Airbnb Names')
sns.barplot(x='Words', y='Count', data=words_df, ax=ax)
plt.xticks(rotation=80)
plt.savefig('Top 25 Words Used In Bangkok AirBnb Names.png',dpi=800,bbox_inches='tight')
plt.show()
```

Interpretasi: Usually host AirBnb in Bangkok use ‘bts, room, near, Bangkok, Sukhumvit, bedroom’ in their title names. From the word we can know where is the place of host AirBnb and what they want to show to customer.
